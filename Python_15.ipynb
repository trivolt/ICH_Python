## insert()

это  метод, который позволяет вставить элемент на указанную позицию в списке, сдвигая остальные элементы вправо.

```python
my_list = [1, 2, 3, 4]
my_list.insert(2, 5)  # Вставка числа 5 на позицию 2
print(my_list)  # [1, 2, 5, 3, 4]
```

## remove()

это  метод, который удаляет первое вхождение указанного элемента из списка. Если элемент не найден, возникает ошибка.

```python
my_list = [1, 2, 3, 4, 2]
my_list.remove(2)
print(my_list)
```

```python
my_list.remove(2)
print(my_list)
```

```python
names = ['Tom','Bob','Yan']
print(names.remove('Joe'))
```

# Кортежи (tuple)

Кортеж - это неизменяемая коллекция, которая похожа на список, но не может быть изменена после создания. Кортежи определяются с помощью круглых скобок и могут содержать элементы разных типов данных.

```python
my_tuple = (1, "two", 3.0)
print(my_tuple)
```

```python
salary_tuple = (1000, 1200, 1300, 900, 800)
#type(salary_tuple)
type(list(salary_tuple))
```

```python
print(salary_tuple[0])
#salary_tuple
```

```python
# попробуем изменить первое значение кортежа
print(type(salary_tuple))
salary_tuple[0] = 100000
print(salary_tuple)
```

```python
# кортеж из одного элемента задается так:
t = ('one',)
type(t)
```

```python
t = ('one')
type(t)
```

Использование кортежей может быть полезным, когда необходима неизменяемость данных.

Если нужно защитить данные от случайного или намеренного изменения, то можно использовать кортежи. Например, этот тип данных подойдёт для сохранения информации о датах рождения клиентов.

Кортежи занимают в памяти меньше места, чем списки. Это легко проверить:

```python
a = (10, 20, 30, 40, 50, 60) # Кортеж
b = [10, 20, 30, 40, 50, 60] # Список


print(a.__sizeof__()) # Размер кортежа
print(b.__sizeof__()) # Размер списка
```

### Оператор *

Раскрытие последовательности с помощью *:
Оператор * (звездочка) может использоваться для раскрытия последовательности, такой как список или кортеж, в отдельные элементы. Например:


```python
my_list = [1, 2, 3]
print(my_list)  # 1 2 3
```

```python
my_list = [1, 2, 3]
print(*my_list)  # 1 2 3
```
